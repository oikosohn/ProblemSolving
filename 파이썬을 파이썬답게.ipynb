- unpacking: packing과 반대되는 개념인 unpacking이라는 개념이 있습니다. packing은 여러개의 객체를 하나의 객체로 합쳐주었습니다. unpacking은 여러개의 객체를 포함하고 있는 하나의 객체를 풀어줍니다. 함수에서 unpacking을 할때는, 매개변수에서 \*을 붙이는게 아니라 인자 앞에 \*을 붙여서 사용합니다.

출처 : https://wikidocs.net/22801

```
# unpacking
def sum(a, b, c):
    return a + b + c

numbers = [1, 2, 3]
# sum(numbers) # error

print(sum(*numbers)) # 출력 : 6

unpacking 순서
# 1. sum(*numbers)
# 2. sum(*[1, 2, 3])
# 3. sum(1, 2, 3)

```

unpacking은 함수를 호출할때 인자를 해체하는 개념이기 때문에, 해채된 결과가 함수의 매개변수에 갯수와 다르다면 에러가 발생합니다.

### 순열과 조합

https://programmers.co.kr/learn/courses/4008/lessons/13347

```
def permute(arr):
    result = [arr[:]]
    c = [0] * len(arr)
    i = 0
    while i < len(arr):
        if c[i] < i:
            if i % 2 == 0:
                arr[0], arr[i] = arr[i], arr[0]
            else:
                arr[c[i]], arr[i] = arr[i], arr[c[i]]
            result.append(arr[:])
            c[i] += 1
            i = 0
        else:
            c[i] = 0
            i += 1
    return result
```

### 가장 많이 등장하는 알파벳 찾기 - Counter

```
from collections import Counter

my_str = input().strip()
myStr = list(my_str)
myList = Counter(myStr)
# print(mylist)
maxList = list(map(int, myList.values()))
maxList.sort(reverse=True)
maxElement = maxList[0]
answer = []
for k,v in myList.items():
    if maxElement == myList[k]:
        answer.append(k)
    else:
        continue                  
print(''.join(sorted(answer)))
```

```
from collections import Counter
my_str = input().strip()
myStr = sorted(my_str)
answer = []
countStr = Counter(myStr)
cnt = countStr.most_common()
for i in range(len(cnt)):
    if cnt[i][1] is cnt[0][1]:
        answer.append(cnt[i][0])
print(''.join(answer))
```

### 이진 탐색하기 - binary search

이진 탐색 - 오름차순으로 정렬된 리스트에서 특정한 값의 위치를 찾는 알고리즘

https://programmers.co.kr/learn/courses/4008/lessons/13173

```
def bisect(a, x, lo=0, hi=None):
    if lo < 0:
        raise ValueError('lo must be non-negative')
    if hi is None:
        hi = len(a)
    while lo < hi:
        mid = (lo + hi) // 2
        if a[mid] < x:
            lo = mid + 1
        else:
            hi = mid
    return lo

mylist = [1, 2, 3, 7, 9, 11, 33]

# 3의 인덱스 리턴
print(bisect(mylist, 3))
```

```
import bisect
mylist = [1, 2, 3, 7, 9, 11, 33]
# 3 다음에 있는 insertion point 리턴
print(bisect.bisect(mylist, 3))

# 3의 인덱스 리턴
print(bisect.bisect(mylist, 3)-1)
```

```
class Coord(object):
    def __init__ (self, x, y):
        self.x, self.y = x, y

point = Coord(1, 2)
print( '({}, {})'.format(point.x, point.y) ) 

# 또는
def print_coord(coord):
    print( '({}, {})'.format(coord.x, coord.y) )
print_coord(point)
```

```
class Coord(object):
    def __init__ (self, x, y):
        self.x, self.y = x, y
    def __str__ (self): # __str__ 메소드 :  class 내부에서 출력 format을 지정
        return '({}, {})'.format(self.x, self.y)

point = Coord(1, 2)
print(point)
```
